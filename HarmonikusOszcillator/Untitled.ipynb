```python
def analX(om,t,x0,v0):
    return x0*cos(om*t) + v0/om*sin(om*t)

def analV(om,t,x0,v0):
    return -x0*om*sin(om*t) + v0*cos(om*t)

t = linspace(0,30,100)
```
