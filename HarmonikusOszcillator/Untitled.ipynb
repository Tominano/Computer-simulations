```python
def analX(om,t,x0,v0):
    return x0*cos(om*t) + v0/om*sin(om*t)

def analV(om,t,x0,v0):
    return -x0*om*sin(om*t) + v0*cos(om*t)

t = linspace(0,300,1000)
```

```python
figure(figsize = (15,15))
title("Fázistér", fontsize = 30)
plot(analX(0.5,t,0,1),analV(0.5,t,0,1), label = "analitikus 0.5$\omega$")
plot(dat_1[:,1],dat_1[:,2], label = "3$\omega$")#3
plot(dat_2[:,1],dat_1[:,2], label = "1$\omega$")#1
plot(dat_3[:,1],dat_1[:,2], label = "2$\omega$")#2
plot(dat_4[:,1],dat_1[:,2], label = "4$\omega$")#4
plot(dat_5[:,1],dat_1[:,2], label = "5$\omega$")#5
plot(dat_6[:,1],dat_1[:,2], label = "6$\omega$")#6
plot(dat_7[:,1],dat_1[:,2], label = "0.5$\omega$")#0.5
xlabel('Idő', fontsize = 30)
ylabel('Sebesség', fontsize = 30)
axvline(x=0, color = 'r')
axhline(y=0, color = 'r')
legend(loc='upper left',fontsize=15)
grid(True)
```
