```python
def ada_cross(ada):
    te = zeros((100,3))
    t = linspace(0,100,101)
    i = 0
    for j in range(0,len(ada),1):
        if ((ada[j,0]-t[i] > 1e-26)):
            te[i] = ada[j,:]
            i+=1
    return te
```

```python
ada0=ada_cross(dat_earth_ark_dt0001__acc_e_6)
ada1=ada_cross(dat_earth_ark_dt0001__acc_e_8)
ada2=ada_cross(dat_earth_ark_dt001__acc_e_10)
ada3=ada_cross(dat_earth_ark_dt0001__acc_e_12)
```
